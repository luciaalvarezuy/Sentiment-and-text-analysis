# Example: calculating the frequency of words of a song

Sonf: Perfect
    
Autfor: Ed Sheeran
    
Text extracted from: https://www.letras.com/ed-sheeran/perfect/


```python
original_text="""I found a love for me.
Darling, just dive right in and follow my lead.
I found a girl, beautiful and sweet.
I never knew you were the someone waiting for me.

'Cause we were just kids when we fell in love.
Not knowing what it was.
I will not give you up this time.
Darling, just kiss me slow.
Your heart is all I own.
And in your eyes, you're holding mine.
"""
```

```python
#Divide text into words and omit comma in word count
tokenizer = nltk.RegexpTokenizer(r"\w+")
tokenized_word = tokenizer.tokenize(original_text)
print(tokenized_word)
```

```python
#Print the frequency distribution of the text
fdist = FreqDist(tokenized_word)
print(fdist)
```

Diferentes métodos que permiten determinar rápidamente las palabras de uso frecuente en el texto.

I.Ver las dos palabras más comunes

```python
fdist.most_common(2)
```

En el texto "I" es la primera palabra que se repite con mayor frecuencia, específicamente se repite 5 veces. Luego, "me" es la segunda palabra con mayor repetición, la cual se repite 3 veces


II. Ver las 10 palabras más repetidas y sus respectivas frecuencias en una tabla:


```python
fdist.tabulate(10)
```

Previamente, se observan las 10 palabras con mayor frecuencia


III. Gráfico de distribución de frecuencias

```python
plt.figure(figsize=(14,12))
fdist.plot(54,cumulative=False)
plt.show()
```

Descripción de los datos:

```python
Sentiment_count=df.groupby('airline_sentiment').count()
plt.bar(Sentiment_count.index.values, Sentiment_count['text'])
plt.xlabel('Review Sentiments')
plt.ylabel('Number of Review')
plt.show()
```

```python
#Generación de características mediante una bolsa de palabras

#tokenizer para eliminar los elementos no deseados de los datos, como los símbolos y los números

token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(lowercase=True,stop_words='english',ngram_range = (1,1),tokenizer = token.tokenize)
text_counts= cv.fit_transform(df['text'])

#Dividamos el conjunto de datos usando la función train_test_split()
X_train, X_test, y_train, y_test = train_test_split(
    text_counts, df['airline_sentiment'], test_size=0.3, random_state=1)

# Generación de modelos mediante Multinomial Naive Bayes
clf = MultinomialNB().fit(X_train, y_train)
predicted= clf.predict(X_test)
print("Precisión MultinomialNB :",metrics.accuracy_score(y_test, predicted))
```

Se obtuvo una tasa de clasificación del 74.91%.

```python
#Generación de características mediante TF-IDF
tf=TfidfVectorizer()
text_tf= tf.fit_transform(df['text'])

X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(
    text_tf, df['airline_sentiment'], test_size=0.3, random_state=123)

# Model Generation Using Multinomial Naive Bayes
clf_2 = MultinomialNB().fit(X_train_2, y_train_2)
predicted_2= clf_2.predict(X_test_2)
print("MultinomialNB Accuracy:",metrics.accuracy_score(y_test_2, predicted_2))
```

Se obtuvo una tasa de clasificación del 67.55%

```python
# Eliminar enlaces. La primera operación es eliminar todos los enlaces, ya que no tienen ningún significado predictiv
df["clean_text"] = df["text"].apply(lambda s: ' '.join(re.sub("(w+://S+)", " ", s).split()))
df[["text","clean_text"]]
```

```python
# Eliminar puntuaciones
df["clean_text"] = df["clean_text"].apply(lambda s: ' '.join(re.sub("[.,!?:;-='...@#_]", " ", s).split()))
df[["text","clean_text"]]
```

```python
# Eliminar emojis

def deEmojify(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')
df["clean_text"] = df["clean_text"].apply(lambda s: deEmojify(s))
df[["text","clean_text"]]
```

```python
# Eliminar las palabras vacías
stop = set(stopwords.words('english'))

def rem_en(input_txt):
    words = input_txt.lower().split()
    noise_free_words = [word for word in words if word not in stop_words] 
    noise_free_text = " ".join(noise_free_words) 
    return noise_free_text

df["clean_text"] = df["clean_text"].apply(lambda s: rem_en(s))
df[["text","clean_text"]]
```

```python
#Copiar las dos columnas en un nuevo dataframe
df_subset = df[['tweet_id', 'clean_text']].copy()
df_subset
```

####  Para finalizar este ejercicio, se genera puntuaciones de polaridad de los sentimientos y se realiza su visualización

```python
# Dataframe vacio para luego ingresar datos:
df1=pd.DataFrame()
df1['tweet_id']=['99999999999']
df1['sentiment_type']='NA999NA'
df1['sentiment_score']=0
```

En este análisis solamente se va a observar "compound" dado que es una puntuación global que combina los sentimientos negativos, positivos y neutros en una única puntuación.

```python
print('Procesamiento del análisis de sentimientos...')
sid = SentimentIntensityAnalyzer()
t_df = df1
for index, row in df_subset.iterrows():
    scores = sid.polarity_scores(row[1])
    for key, value in scores.items():
        temp = [key,value,row[0]]
        df1['tweet_id']=row[0]
        df1['sentiment_type']=key
        df1['sentiment_score']=value
        t_df=t_df.append(df1)
#eliminar la fila ficticia con tweet_id  = 99999999999
t_df_cleaned = t_df[t_df.tweet_id != '99999999999']

#eliminar los duplicados si existen
t_df_cleaned = t_df_cleaned.drop_duplicates()

# conservar sólo las filas en las que sentiment_type = compound

t_df_cleaned = t_df[t_df.sentiment_type == 'compound']
print(t_df_cleaned.head(10))
```

```python
#Combinar los datos obtenidos con la tabla previa
df_output = pd.merge(df, t_df_cleaned, on='tweet_id', how='inner')
print(df_output.head(10))
```
